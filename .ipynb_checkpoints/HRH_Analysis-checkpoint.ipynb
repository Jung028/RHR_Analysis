### **Predictive Analysis**  
- üî¢ Uses **Linear Regression** to predict future **Avg BPM (Resting Heart Rate)**.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from datetime import datetime


# Load data from CSV file
file_path = r"Training_Data - Cycling.csv"
df = pd.read_csv(file_path)
```


### 1. Resting Heart Rate Trend
#### Resting Heart Rate Trend
This graph plots your average BPM over time to determine if your resting heart rate is decreasing. A lower resting heart rate over time typically indicates improved cardiovascular fitness and increased efficiency of the heart.


```python
# Convert to DataFrame
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%y')

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Avg BPM'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Average BPM')
plt.title('Resting Heart Rate Trend')
plt.grid()
plt.show()
```


### 2. Power-to-Heart Rate Efficiency
#### Power-to-Heart Rate Efficiency
This scatter plot shows the relationship between power output (watts) and average BPM. If your heart rate remains stable while power output increases, it indicates better efficiency and improved endurance.


```python
plt.figure(figsize=(10,5))
plt.scatter(df['Avg BPM'], df['Avg Watt'], c='blue', label='Avg Watt')
plt.scatter(df['Avg BPM'], df['Max Watt'], c='red', label='Max Watt')
plt.xlabel('Average BPM')
plt.ylabel('Power Output (Watts)')
plt.title('Power-to-Heart Rate Efficiency')
plt.legend()
plt.grid()
plt.show()
```

### 3. Cadence vs. Speed Relationship

#### Cadence vs. Speed Relationship
This graph examines how cadence (RPM) affects your speed (MPH). A higher cadence at a given speed suggests improved pedaling efficiency, while a declining trend might indicate fatigue or inefficiencies in technique.


```python

plt.figure(figsize=(10,5))
plt.scatter(df['Avg RPM'], df['Avg Speed (MPH)'], c='green')
plt.xlabel('Average RPM')
plt.ylabel('Average Speed (MPH)')
plt.title('Cadence vs. Speed Relationship')
plt.grid()
plt.show()
```

### 4. Heart Rate vs. Speed Relationship

#### Heart Rate vs. Speed Relationship
This plot compares your average BPM to speed (MPH). If you can maintain the same speed at a lower heart rate over time, it signifies better endurance and cardiovascular conditioning.


```python

plt.figure(figsize=(10,5))
plt.scatter(df['Avg BPM'], df['Avg Speed (MPH)'], c='purple')
plt.xlabel('Average BPM')
plt.ylabel('Average Speed (MPH)')
plt.title('Heart Rate vs. Speed Relationship')
plt.grid()
plt.show()
```

### 5. Peak Power Trend
#### Peak Power Trend
This graph tracks maximum wattage over time to assess improvements in strength and power output. An upward trend suggests increasing leg strength and muscular endurance, key for sustaining higher efforts during cycling sessions.


```python
plt.figure(figsize=(10, 5))
plt.bar(df['Date'], df['Max Watt'], color='blue')
plt.xlabel('Date')
plt.ylabel('Max Watt')
plt.title('Peak Power Distribution')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```


### 6. Resting Heart Rate Prediction
#### Performance Prediction
- lower RHR means less effort required to perform the same task, while higher RHR means higher effort.

Formula :

##### RHR = Avg¬†BPM ‚àí (Calories¬†Burned / 100)
 
Why This Approximation?
RHR is typically measured in the morning before activity, but since we don‚Äôt have actual RHR data, I estimated it from Avg BPM.
Higher calorie burn means a more intense workout, which often leads to temporary heart rate elevation but can lower RHR over time due to improved fitness.
Subtracting a scaled calorie value (dividing by 100) creates a reasonable trend where increased effort correlates with improved fitness and lower RHR.

```python


# Load dataset
file_path = r"Training_Data - Cycling.csv"
df = pd.read_csv(file_path)

# Convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%y")

# Define Resting Heart Rate (RHR) approximation
# Assuming RHR is lower when recovery is better. Using Avg BPM trends for estimation.
df['RHR'] = df['Avg BPM'] - (df['Calories Burned'].str.replace(' kcal', '').astype(int) / 100)

# Higher effort indicator (if calories burned > 500, consider it high effort)
df['Effort'] = df['Calories Burned'].str.replace(' kcal', '').astype(int) > 500

# Prepare data for regression
df = df.sort_values(by="Date")
X = np.array((df['Date'] - df['Date'].min()).dt.days).reshape(-1, 1)
y = df['RHR'].values.reshape(-1, 1)

# Train linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict future RHR
future_dates = np.array([(df['Date'].max() + pd.Timedelta(days=i) - df['Date'].min()).days for i in range(1, 31)]).reshape(-1, 1)
predicted_rhr = model.predict(future_dates)

# Plot actual vs predicted RHR
plt.figure(figsize=(10, 5))
plt.scatter(df['Date'], df['RHR'], label="Actual RHR", color='blue')
plt.plot(df['Date'], model.predict(X), label="Regression Line", color='red')
plt.plot(df['Date'].max() + pd.to_timedelta(range(1, 31), unit='D'), predicted_rhr, label="Predicted RHR", linestyle="dashed", color="green")
plt.xlabel("Date")
plt.ylabel("Resting Heart Rate (RHR)")
plt.title("Predicted Future RHR Trend")
plt.legend()
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
